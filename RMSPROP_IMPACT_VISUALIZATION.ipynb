```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rmsprop_lr_optimizer(gradients,lr=0.1,beta=0.9,epsilon=1e-8):
    theta=5 #Here we are defining the intial weight theta
    E_g2=0 #Moving average of squared gradients
    theta_values_rmsprop=[]
    for g in gradients:
        E_g2=beta*E_g2+(1-beta)*g**2
        theta=theta-lr/(np.sqrt(E_g2)+epsilon)*g
        theta_values_rmsprop.append(theta)
    
    return theta_values_rmsprop
```

```python
gradients=[3,-2,4,-3,5]
updates_rmsprop=rmsprop_lr_optimizer(gradients)
```

```python
def grad_without_rmsprop(gradients,lr=0.1):
    theta_xr=5
    theta_values_x_rmsprop=[]
    for g in gradients:
        theta_xr=theta_xr-lr*g
        theta_values_x_rmsprop.append(theta_xr)

    return theta_values_x_rmsprop
```

```python
updates_x_rmsprop=grad_without_rmsprop(gradients)
```

```python
plt.figure(figsize=(8,5))
plt.plot(updates_rmsprop,marker="o",linestyle="-",label="Updates with RMSPROP")
plt.plot(updates_x_rmsprop,marker="s",linestyle="--",label="Updates without RMSPROP")
plt.title("Visualizing RMSPROP")
plt.xlabel("Iterations")
plt.ylabel("Values")
plt.legend()
plt.grid()
plt.show()
```
