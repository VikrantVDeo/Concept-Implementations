```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sgd_momentum(gradients,lr=0.1,beta=0.9):
    v=0 #This is my initial velocity term
    theta=0 #These are my weights
    theta_values=[]

    for g in gradients:
        v=beta*v+(1-beta)*g
        theta=theta-lr*v
        theta_values.append(theta)
    
    return theta_values
    
```

```python
def sgd_without_momentum(gradients,lr=0.1,beta=0.9):
    theta_xm=0
    theta_xm_values=[]
    for g in gradients:
        theta_xm=theta_xm-lr*g
        theta_xm_values.append(theta_xm)
    return theta_xm_values
```

```python
gradients=[3,-2,4,-3,5]
updatesm=sgd_momentum(gradients)
updateswm=sgd_without_momentum(gradients)
```

```python
plt.figure(figsize=(8,5))
plt.plot(updatesm,marker="o",linestyle="-", label="Updates with Momentum")
plt.plot(updateswm, marker="s",linestyle="--",label="Updates without Momentum")
plt.xlabel("Iteration")
plt.ylabel("Value")
plt.title("Visualizing the effect of Momentum")
plt.legend()
plt.grid()
plt.show()
```
